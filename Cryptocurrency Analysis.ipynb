```python
#importing necessary libraries
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#creating an array of coins
cryptocurrencies=['DOGE-USD','BTC-USD','ETH-USD','XRP-USD']
```

```python
#getting data from  yfinance
data=yf.download(cryptocurrencies,start='2020-01-01',end='2022-11-07')
data.head()
```

```python
#adj close
adj_close=data['Adj Close']
adj_close.head()
```

```python
#percentage change in the adjusted closing prices
returns=adj_close.pct_change().dropna(axis=0)
returns
```

```python
#creating subplots of the coins
fig,axs=plt.subplots(2,2,figsize=(15,5),gridspec_kw={'wspace': 0.1,'hspace': 0.7})
axs[0,0].plot(adj_close['DOGE-USD'])
axs[0,0].set_title('DOGE')
axs[0,1].plot(adj_close['BTC-USD'])
axs[0,1].set_title('BTC')
axs[1,0].plot(adj_close['ETH-USD'])
axs[1,0].set_title('ETH')
axs[1,1].plot(adj_close['XRP-USD'])
axs[1,1].set_title('XRP')
plt.show()
```

```python
#creating subplots of the coins
fig,axs=plt.subplots(2,2,figsize=(15,5),gridspec_kw={'wspace': 0.1,'hspace': 0.7})
axs[0,0].plot(returns['DOGE-USD'])
axs[0,0].set_title('Returns on DOGE')
axs[0,1].plot(returns['BTC-USD'])
axs[0,1].set_title('Returns on BTC')
axs[1,0].plot(returns['ETH-USD'])
axs[1,0].set_title('Returns on ETH')
axs[1,1].plot(returns['XRP-USD'])
axs[1,1].set_title('Returns on XRP')
plt.show()
```

```python
#creating subplots of the coins
fig,axs=plt.subplots(2,2,figsize=(15,5),gridspec_kw={'wspace': 0.1,'hspace': 0.7})
axs[0,0].hist(returns['DOGE-USD'],bins=100,range=(-0.5,0.5))
axs[0,0].set_title(' Histogram of Returns on DOGE')
axs[0,1].hist(returns['BTC-USD'],bins=100,range=(-0.5,0.5))
axs[0,1].set_title('Histogram of Returns on BTC')
axs[1,0].hist(returns['ETH-USD'],bins=100,range=(-0.5,0.5))
axs[1,0].set_title('Histogram of Returns on ETH')
axs[1,1].hist(returns['XRP-USD'],bins=100,range=(-0.5,0.5))
axs[1,1].set_title('Histogram of Returns on XRP')
plt.show()
```

```python
#checking correlations between the coins
corr=returns.corr()
sns.heatmap(corr,annot=True,cmap='Blues')
```
